```python
import pandas as pd
import numpy as np

dt1 = pd.read_csv('/content/drive/MyDrive/Documentos/file 1.csv')

dt2 = pd.read_excel('/content/drive/MyDrive/Documentos/Consolidado Cupos Maximos 1.xlsx')

#Cambiamos la tipografia del nombre de columnas
dt2.columns = dt2.columns.str.upper()
dt1.columns = dt1.columns.str.upper()

# Aplicar metodo strip a cada registro del dataframe
dt1 = dt1.applymap(lambda x: x.strip() if isinstance(x, str) else x)
dt2 = dt2.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Cantidad de registros por dataframe
print(f'Registros del fichero "file 1.csv": {len(dt1)} y el fichero "Consolidado Cupo Maximos 1.xlsx": {len(dt2)}')
```

```python
# Verificar longitud registro patentes
lista_parcial_dt2 = dt2['PATENTE'].dropna().apply(len)
lista_parcial_dt1 = dt1['PATENTE'].dropna().apply(len)

# Encuentra la longitud máxima
print(f'Longitud de registro patente "file 1.csv": [{min(lista_parcial_dt2),max(lista_parcial_dt2)}]')
print(f'Longitud de registro patente "Consolidado cuypo maximos 1.xlsx: [{min(lista_parcial_dt1), max(lista_parcial_dt1)}]')
```

```python
def quitar_codigo_verificador(cadena):
    if len(cadena) == 9:
        return cadena[:-2]
    else:
        return cadena

dt1["PATENTE"] = dt1["PATENTE"].apply(quitar_codigo_verificador)
```

```python
# Cantidad de registros por dataframe
print(f'Registros del fichero "file 1.csv": {len(dt1)} y el fichero "Consolidado Cupo Maximos 1.xlsx": {len(dt2)}')
```

```python
dt_merge = pd.merge( dt1,dt2, on = 'PATENTE',how = 'left')
dt_merge.sample(5)
```

```python
# Cantidad de registros por dataframe
print(f'Registros del dataframe en cruce: {len(dt_merge)} ')
```

```python
# Patentes nulas en el Dataframe Merge
cantidad_nulos = dt_merge['PATENTE'].isnull().sum()
cantidad_nulos
```

```python
# Impresiones de ficheros
dt_merge.to_excel('/content/drive/MyDrive/Documentos/merge.xlsx', index=False)
dt_merge.to_csv('/content/drive/MyDrive/Documentos/merge.csv', index=False)
```
