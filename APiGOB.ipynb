```python
import pandas as pd
import uuid

# Rutas

objetivos_basica = '/content/drive/MyDrive/Taller de Desarrollo Soluciones 2/APIS_APP/OBJETIVOS_BASICA.xlsx'

INSERTS_UNIDADES = '/content/drive/MyDrive/Taller de Desarrollo Soluciones 2/APIS_APP/INSERTS_UNIDADES.xlsx'
INSERTS_NIVEL = '/content/drive/MyDrive/Taller de Desarrollo Soluciones 2/APIS_APP/INSERTS_NIVEL.xlsx'
INSERTS_ASIGNATURAS = '/content/drive/MyDrive/Taller de Desarrollo Soluciones 2/APIS_APP/INSERTS_ASIGNATURAS.xlsx'
INSERTS_OBJETIVOS = '/content/drive/MyDrive/Taller de Desarrollo Soluciones 2/APIS_APP/INSERTS_OBJETIVOS.xlsx'

```

## 2. Recursos y diccionarios

```python
DF_OBJETIVOS = pd.read_excel(objetivos_basica)
OBJETIVOS = DF_OBJETIVOS['descObjetivo']
ASIGNATURAS = DF_OBJETIVOS['nombreAsignatura'].unique()
NIVELES = DF_OBJETIVOS['nombreNivel'].unique()
UNIDADES = DF_OBJETIVOS['nombreUnidad'].unique()

# Definir una función para extraer las primeras dos letras de cada palabra y el primerconjunto id
def extract_initials(text,separator):

  if separator == '-':
    segmentos = text.split('-')
    return segmentos[0]
  elif separator == ' ':
    palabras = text.split()
    slug = palabras[0][0:3]
    return slug
```

## 3. Asignar UUIID4 unico a cada asignatura

```python
ASIGNATURAS_ARRAY = []
for asignatura in ASIGNATURAS:
  ASIGNATURAS_ARRAY.append([asignatura,str(uuid.uuid4())])

ASIGNATURAS_DF = pd.DataFrame(ASIGNATURAS_ARRAY,columns=['nombre','idAsignatura'])

ASIGNATURAS_DF['slug'] = ASIGNATURAS_DF['nombre'].apply(extract_initials, separator = ' ') + ASIGNATURAS_DF['idAsignatura'].apply(extract_initials, separator= '-')
ASIGNATURAS_DF.rename(columns = {'nombre': 'nombreAsignatura'}, inplace=True)

ASIGNATURAS_DF
```

## 4. Asignar UUIID4 unico a cada nivel


```python
NIVELES_ARRAY = []
for nivel in NIVELES:
  NIVELES_ARRAY.append([nivel,str(uuid.uuid4())])

NIVELES_DF = pd.DataFrame(NIVELES_ARRAY,columns=['nombre','idNivel'])
NIVELES_DF['slug'] = NIVELES_DF['nombre'].apply(extract_initials, separator = ' ') + NIVELES_DF['idNivel'].apply(extract_initials, separator= '-')
NIVELES_DF.rename(columns = {'nombre': 'nombreNivel'}, inplace=True)

NIVELES_DF
```

## 5. Asignar UUIID4 a cada unidad


```python
UNIDAD_ARRAY = []
for unidad in UNIDADES:
  UNIDAD_ARRAY.append([unidad,str(uuid.uuid4())])

UNIDADES_DF = pd.DataFrame(UNIDAD_ARRAY,columns=['nombre','idUnidad'])
UNIDADES_DF['slug'] = UNIDADES_DF['nombre'].apply(extract_initials, separator = ' ') + UNIDADES_DF['idUnidad'].apply(extract_initials, separator= '-')
UNIDADES_DF['idAsignatura'] = ASIGNATURAS_DF[ASIGNATURAS_DF['nombreAsignatura'] == 'Matemática'].iloc[0, 1]
UNIDADES_DF.rename(columns = {'nombre': 'nombreUnidad'}, inplace=True)


UNIDADES_DF
```

## 6. Asignar UUIID4 a cada objetivo

```python
OBJETIVO_ARRAY = []
for objetivo in OBJETIVOS:
  OBJETIVO_ARRAY.append([objetivo,str(uuid.uuid4())])

OBJETIVOS_DF = pd.DataFrame(OBJETIVO_ARRAY,columns=['nombre','idObjetivo'])
OBJETIVOS_DF['slug'] = OBJETIVOS_DF['nombre'].apply(extract_initials, separator = ' ') + OBJETIVOS_DF['idObjetivo'].apply(extract_initials, separator= '-')
OBJETIVOS_DF.rename(columns = {'nombre': 'descObjetivo'}, inplace=True)


OBJETIVOS_DF
```

## Unificar columnas

```python
DF_OBJETIVOS = DF_OBJETIVOS.merge(NIVELES_DF[['nombreNivel','idNivel']], how='inner', left_on='nombreNivel', right_on='nombreNivel')
DF_OBJETIVOS.sample(2)
```

```python
DF_OBJETIVOS = DF_OBJETIVOS.merge(OBJETIVOS_DF[['descObjetivo','idObjetivo','slug']], how='inner', on='descObjetivo')

DF_OBJETIVOS.sample(2)
```

```python
DF_OBJETIVOS = DF_OBJETIVOS.merge(UNIDADES_DF[['nombreUnidad','idUnidad']], how='inner', on='nombreUnidad')

DF_OBJETIVOS
```

### Matriz

```python
# Crear listas para almacenar datos
paises = []
regiones_list = []
comunas_list = []


# Iterar sobre la lista de localidades
for localidad in localidades:
    pais = localidad['country']

    for region in localidad['regiones']:
        region_name = region['name']
        comunas = region['comunas']

        # Añadir datos a las listas
        paises.extend([pais] * len(comunas))
        regiones_list.extend([region_name] * len(comunas))
        comunas_list.extend(comunas)

# Crear el DataFrame
localidades_df = pd.DataFrame({
    'Pais': paises,
    'Region': regiones_list,
    'Comuna': comunas_list
})

# Mostrar el DataFrame
localidades_df.sample(5)
```

### Creacion Dataframes y Slugs de Pais, Region y Comuna


```python
id_paises = [str(uuid.uuid4()) for pais in localidades_df['Pais'].unique()]
id_regiones = [str(uuid.uuid4()) for region in localidades_df['Region'].unique()]
id_comuna = [str(uuid.uuid4()) for comuna in localidades_df['Comuna'].unique()]

# Funcion de creacion slug
def extract_initials(text,separator):

  if separator == '-':
    segmentos = text.split('-')
    return segmentos[0]
  elif separator == ' ':
    palabras = text.split()
    slug = palabras[0][0:3]
    return slug


# Crear un DataFrame para mapear países a IDs
paises_df = pd.DataFrame({
    'Pais': localidades_df['Pais'].unique(),
    'ID_Pais': id_paises,
})

paises_df['slug_pais'] = paises_df['Pais'].apply(extract_initials, separator = ' ') + paises_df['ID_Pais'].apply(extract_initials, separator= '-')


regiones_df = pd.DataFrame({
    'Region': localidades_df['Region'].unique(),
    'ID_Region': id_regiones
})

regiones_df['slug_region'] = regiones_df['Region'].apply(extract_initials, separator = ' ') + regiones_df['ID_Region'].apply(extract_initials, separator= '-')

comunas_df = pd.DataFrame({
    'Comuna': localidades_df['Comuna'].unique(),
    'ID_Comuna': id_comuna
})

comunas_df['slug_comuna'] = comunas_df['Comuna'].apply(extract_initials, separator = ' ') + comunas_df['ID_Comuna'].apply(extract_initials, separator= '-')


```

### Merge de Dataframes

```python
localidades_df = pd.merge(localidades_df, paises_df, on='Pais')
localidades_df = pd.merge(localidades_df, regiones_df, on='Region')
localidades_df = pd.merge(localidades_df, comunas_df, on='Comuna')


localidades_df.sample(5)
```

### Sentencias Insert de Pais, Region y Comuna

```python
# Crear SQL_INSERT Pais
for fila in range(len(localidades_df)):
  localidades_df.loc[fila,'SQL_PAIS'] = f"INSERT INTO TESTCREATES_PAIS(NOMBRE_PAIS,ID_PAIS,SLUG) VALUES ('{localidades_df.loc[fila, 'ID_Pais']}','{localidades_df.loc[fila, 'Pais']}','{localidades_df.loc[fila, 'slug_pais']}');"

# Crear SQL_INSERT Region
for fila in range(len(localidades_df)):
  localidades_df.loc[fila,'SQL_REGION'] = f"INSERT INTO TESTCREATES_REGION(ID_REGION,NOMBRE_REGION,PAIS_ID,SLUG) VALUES ({localidades_df.loc[fila, 'ID_Region']}','{localidades_df.loc[fila, 'Region']}','{localidades_df.loc[fila, 'ID_Pais']}','{localidades_df.loc[fila, 'slug_region']}');"

# Crear SQL_INSERT Comuna
for fila in range(len(localidades_df)):
  localidades_df.loc[fila,'SQL_COMUNA'] = f"INSERT INTO TESTCREATES_COMUNA(ID_COMUNA,NOMBRE_COMUNA,REGION_ID,SLUG) VALUES ('{localidades_df.loc[fila, 'ID_Comuna']}','{localidades_df.loc[fila, 'Comuna']}','{localidades_df.loc[fila, 'ID_Region']}','{localidades_df.loc[fila, 'slug_comuna']}');"


localidades_df.sample(5)
```
