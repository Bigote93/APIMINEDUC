## Carga de objetivos en dataframe

```python

# Leer el archivo excel de los objetivos de enseñanza basica
objetivos_basica = '/content/drive/MyDrive/Taller de Desarrollo Soluciones 2/APIS_APP/API_OBJ_BASICA.xlsx'
objetivos_insert = '/content/drive/MyDrive/Taller de Desarrollo Soluciones 2/APIS_APP/API_OBJ_BASICA_inserts.xlsx'

# Lista de objetivos
cursos = [
    'Primero_basico','Segundo_basico','Tercero_basico','Cuarto_basico',
    'Quinto_basico','Sexto_basico','Septimo_basico','Octavo_basico'
    ]

df_columnas = pd.DataFrame(
    columns=['index','idNivel','nombreNivel','descripcionNivel','idAsignatura',
             'nombreAsignatura','idUnidad','nombreUnidad','idObjetivo','descObjetivo'
             ]
    )

# Recorremos las hojas y almacenamos todos los objetivos en un solo dataframe de objetivos basica
for curso in cursos:
  df_objetivos = pd.read_excel(objetivos_basica, sheet_name=curso)
  # Unimos los dataframes en un solo dataframe
  df_columnas = pd.concat([df_columnas,df_objetivos]).reset_index(drop=True)

df_objetivos_basica = df_columnas.drop(columns=['index'])

# Incorporar la columna de insert en codigo SQL

# Eliminar los saltos de linea de la columna descripcion
df_objetivos_basica['descObjetivo'] = df_objetivos_basica['descObjetivo'].str.replace('\n', '')

# Crear una nueva columna con el formato deseado
df_objetivos_basica['sentencia_sql'] = df_objetivos_basica.apply(lambda row: f"INSERT INTO TESTCREATES_OBJETIVOAPRENDIZAJE (id_objetivo_aprendizaje,enunciado_objetivo_aprendizaje,unidad_id) VALUES ('{uuid.uuid4()}','{row['descObjetivo']}', '{row['idUnidad']}');", axis=1)

# Guardamos los elementos en un nuevo dataframe
df_objetivos_basica.to_excel(objetivos_insert, sheet_name='Objetivos_insert')


df_objetivos_basica.sample(5)
```
