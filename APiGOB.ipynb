## (1) Extraer los objetivos de primero basico

Primero contamos cuantos objetivos contiene el primero basico efectivamente y sus respectivos nombres:

* Cantidad de objetivos: Almacenaremos en objetivos_1B
* Titulos de objetivos: Almacenaremos en objetivos_1B_txt

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd

objetivos_1B = 0
objetivos_1B_txt = []

# Obtenemos el contenido de la página web
response = requests.get('https://www.curriculumnacional.cl/portal/Ejes/Matematica/')

# Creamos un objeto BeautifulSoup para analizar el contenido HTML
soup = BeautifulSoup(response.content, 'html.parser')

#Buscamos el contenedor de todos los acoordeones
elemento = soup.find('div', class_ = 'grupo ejes-acordeon acordeon-recuadros panel-group acordeon-izquierda margen-abajo-md')

#Buscamos todos los elementos internos de cada acordeon que contiene cada titulo objetivo como primer elemento
elemnto_internon1 = elemento.find_all('div')

#Barrimos cada elemento de los div anteriores para extraer cada contenedor de nombre objetivo
for i in elemnto_internon1:
  #Extraemos el primer elemento interno que contiene a cada contenedor de titulo pbjetivo
  #Debemos ademas extraer el enlace que contiene el titulo
  enlaces_titulo_obj = i.find_next().find('a')

  #Barrimos para que no sea None debido a que algunos internos no tienen titulo o vacio
  if enlaces_titulo_obj is not None:

    #Barrimos apra que tambien eliminemos los enlaces de Ver recursos
    if enlaces_titulo_obj.text != 'Ver recursos':

      #Teniendo el titulo objetivo solo tomamos en cuenta a aquellos de la forma: MA01 0A XX
      if enlaces_titulo_obj.text.startswith('MA01 OA '):
        objetivos_1B_txt.append(enlaces_titulo_obj.text.replace(' ','-'))
        objetivos_1B +=1

print(f'Total de objetivos: {objetivos_1B}')
````
